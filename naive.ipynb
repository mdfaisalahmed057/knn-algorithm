```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def bayes(Data,x,col,yescount,nocount):
    xyes=0
    xno=0
    for line in Data:
        if line[-1]=='yes':
           xyes+=1;
        else:
           xno+=1
    pxyes=xyes/yescount
    pxno=xno/nocount
    return pxno,pxyes
def main():
    file="/content/car (2).csv"
    data=pd.read_csv(file)
    display(data)
    temp=[]
    fd=csv.reader(open(file))
    for line in fd:
        temp.append(line)
    Data=temp[1:]
    n=len(Data)
    yescount=0;
    nocount=0
    for line in Data:
        if line[-1]=='yes':
            yescount+=1;
        else:
            nocount+=1
    pyes=yescount/n
    pno=nocount/n
    print("enter your new feature color,type,origin")
    x,y,z=input().split()
    pxyes,pxno=bayes(Data,x,0,yescount,nocount)
    pyyes,pyno=bayes(Data,y,1,yescount,nocount)
    pzyes,pzno=bayes(Data,z,2,yescount,nocount)
    resyes=pxyes*pyyes*pzyes*pyes
    resno=pxno*pyno*pzno*pno
    percentageyes=(resyes/(resyes+resno))*100
    percentageno=(resno/(resyes+resno))*100
    pex=[percentageyes,percentageno]
    label=["car_stolen%","car_not_stolen%"]
    plt.pie(pex,labels=label)
    plt.show()
    print("percentage_yes=",percentageyes,"percentage_no",percentageno)
main()
     
```
